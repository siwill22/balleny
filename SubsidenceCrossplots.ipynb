```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pygmt

df = pd.read_excel('Huppert_data.xlsx', skiprows=1)

df.rename(columns={'Peak Elevation\n(m)':'PeakElevation',
                   'Lat. (°)':'Lat',
                   'Long. (°)':'Long'}, 
          inplace=True)

pd.set_option('display.max_rows', None, 'display.max_columns', None)  # more options can be specified also

#df = df[df.PeakElevation<0.]
#df.reset_index(inplace=True)

df
```

```python
AgeAsFloat = []
for i,row in df.iterrows():
    #print(str(row['Age (Ma)']).split())
    AgeAsFloat.append(float(str(row['Age (Ma)']).split()[0]))
    
print(AgeAsFloat)
```

```python
#print(np.array(df['PeakElevation']))
plt.plot(df['ExpectedSubsidence']*1000.,
         df['PeakElevation']*-1., 'o')
plt.ylim(0,3500.)
plt.xlim(0,3500.)
plt.grid()
plt.show()
```

```python
agefile = './Data/EarthByte_AREPS_Muller_etal_2016_AgeGrid.nc'

res = pygmt.grdtrack(df[['Long','Lat']], grid=agefile, newcolname='SeafloorAge')

plt.plot(res['SeafloorAge'], df['PeakElevation'],'o')
plt.ylim(0,-3500.)
plt.show()
```

```python
plt.plot(res['SeafloorAge']-AgeAsFloat, df['PeakElevation'],'o')
plt.ylim(0,-3500.)
plt.show()
```

```python
def GetExpectedSeafloorDepth(seafloor_ages):
    SteinStein = 365. * np.sqrt(seafloor_ages) + 2600.
    SteinStein[seafloor_ages>20.] = 5651. - 2473 * np.exp(-0.0278 * seafloor_ages[seafloor_ages>20.]) 
    return SteinStein

def GetExpectedSubsidence(seafloor_ages,volcano_ages):
    return GetExpectedSeafloorDepth(seafloor_ages) - GetExpectedSeafloorDepth(seafloor_ages-volcano_ages)

expected_subsidi = GetExpectedSubsidence(np.array(res['SeafloorAge']),AgeAsFloat)

plt.plot(expected_subsidi, -1*df['PeakElevation'],'o')
plt.plot(expected_subsidi[:5], -1*df['PeakElevation'][:5],'ro')
plt.xlabel('Expected Subsidence')
plt.ylabel('Observed Subsidence')
plt.ylim(0,3500.)
plt.xlim(0,3500.)
plt.plot([0,4000],[0,4000],'k')
plt.show()
```
